```python
def getKeyWord(treeElement):
    value = ''
    for element in treeElement:
        
        value += element[0] + ' '
    value = value.rstrip()

    return value

def cleanText(text):
    ctext=[]
    for line in text:
        if "== See also ==" in line:
            break
        if not len(line.strip()) < 1 :
            ctext.append(line)
    return ctext
    
df = pd.DataFrame()

for person in wikiId:
    # read text
    text = readText(person)
    # clean empty lines
    ctext= cleanText(text)
#     ctext=[]
#     for line in text:
#         if not len(line.strip()) < 1 :
#             ctext.append(line)
    # ctext contains a list of paragraphs
    # tokenize each paragraph to obtain a list of sentences
    for line in ctext:
        # returns a list of sentences for each paragraph
        ss=nt.sent_tokenize(line)
        for sentence in ss:
            pEntity = 0
            lEntity = 0
            # obtain the entities for each sentence
            ne_tree = nltk.ne_chunk(pos_tag(word_tokenize(sentence)))
            # extract 2 types of entities for now: PERSON, LOCATION
            keyPersons = []
            keyLocations = []
            for t in ne_tree.subtrees():
                print(t)
                # create a dataset with this information
                if (t.label() == 'PERSON'):
#                     value = ''
#                     for element in t:
#                         value += element[0] + ' '
#                     value = value.rstrip()
#                     print(value)
                    valuePerson = getKeyWord(t)
                    print(valuePerson)
                    keyPersons.append(valuePerson)
                    pEntity = 1
                if t.label() == 'LOCATION':
                    valueLocation = getKeyWord(t)
                    keyLocations.append(valueLocation)
                    lEntity = 1
                    
            datarow = pd.Series(data={'person':person,'sentence':sentence,'peopleEntity':pEntity,'locationEntitiy':lEntity,
                                      'peopleKeyWords':keyPersons,'locationKeyWords':keyLocations})
            df = df.append(datarow,ignore_index=True)
df.info()
```

```python
def extract_NN(sent):
    grammar = r"""NBAR:
    
    {<NN.*>*<NN.*>} # Nouns and Adjectives, terminated with Nouns
    NP:
    {<NBAR>} 
    {<NBAR><IN><NBAR>} # Above, connected with in/of/etc...
    """
    chunker = nltk.RegexpParser(grammar)
    ne = set()
    chunk = chunker.parse(nltk.pos_tag(nltk.word_tokenize(sent)))
    for tree in chunk.subtrees(filter=lambda t: t.label() == 'NP'):
        ne.add(' '.join([child[0] for child in tree.leaves()]))
    return ne
```
